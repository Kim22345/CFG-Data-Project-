```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import csv

#read file 1
df = pd.read_csv('202105_fixed_laua_coverage_r01.csv', header=0)
df.info()
df.head(10)
```

```python
#read file 2
df2 = pd.read_csv('area_grid.csv', header=0)
df2.info()
df2.head(10)
```

```python
#extract data from certain columns
sdf2 = df2[['District', 'District Code', 'Latitude','Longitude', 'Population']]
sdf2.head()
```

```python
#join subdf1 and subdf2 (inner join)
innerjoindf = pd.merge(df,sdf2,left_on ='laua',right_on = 'District Code',how ='inner')
# innerjoindf.shape
innerjoindf.info()
innerjoindf.head()
```

```python
'''
Kim's comment:
One thing highlighted during the pandemic was the importance of high quality connections to enable children to access online learning,
particularly live lessons with their class. For video conferencing, recommended minimum download speeds are around 8MBps 
so I have filtered the dataset to find the areas that do do not meet this (and therefore may not be able to access online learning)'''

new_df = innerjoindf[(innerjoindf['% of premises with 0<2Mbit/s download speed'] >=1) | (innerjoindf['% of premises with 2<5Mbit/s download speed'] >=1) |(innerjoindf['% of premises with 5<10Mbit/s download speed'] >=1)]
new_df
# new_df.info()
```
